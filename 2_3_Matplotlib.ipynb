##<font color='blue'>2.3.2 Figure class</font>

>ใช้กำหนดค่าต่างๆ ในหน้าต่างที่แสดงกราฟ เช่น ขนาด สี ขอบ ของแผนภูมิเป็นต้น

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
```

```python
x = [10, 20, 30, 40]
y = [15, 25, 40, 70]
```

```python
fig = plt.figure(figsize=(10,5), facecolor='pink', edgecolor='aqua', linewidth = 8) #size 10x5
ax = fig.add_axes([0,0,0.5,0.5]) #creating a new axes for the figure [x0,y0,width,height]
ax.plot(x,y) #plotting the data

plt.title("Linear Graph", fontsize = 10, color = "red")
plt.xlabel('X-Axis') #adding label on x-axis
plt.ylabel('Y-Axis') #adding label on y-axis
plt.ylim(0,80) #setting the limit of y-axis
plt.xticks(x,labels=["2010", "2020", "2030", "2040"]) #changing the label of x-axis
plt.legend(["data-1"])
plt.show()
```

```python
fig = plt.figure(figsize=(5,4)) #size 5x4
ax = fig.add_axes([0,0,1,1]) #creating a new axes for the figure [x0,y0,width,height]
ax1 = ax.plot(x,y) #plotting the 1st data
ax2 = ax.plot(y,x) #plotting the 2nd data

ax.set_title("Linear Graph", fontsize = 10, color = "red")
ax.set_xlabel('X-axis') #adding label on x-axis
ax.set_ylabel('Y-axis') #adding label on y-axis

plt.legend(["data-1", "data-2"]) #adding 2 legends
plt.show()
```

```python
#ตัวอย่างการใช้งาน subplot2grid
import matplotlib.pyplot as plt

def annotate_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)

fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 1), rowspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2))
ax4 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
ax5 = plt.subplot2grid((3, 3), (2, 2))

annotate_axes(fig)
plt.show()
```

```python
#ตัวอย่างการใช้งานแผนภูมิแท่ง
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'coconut', 'orange', 'cherry']
counts = [100, 40, 70, 30]
bar_labels = ['red', 'green', 'orange', 'blue']
bar_colors = ['tab:red', 'tab:green', 'tab:orange', 'tab:blue']

ax.bar(fruits,counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(20000000)

# Compute pie slices
N = 10
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
```

```python
#ตัวอย่างการใช้งานฮีสโตแกรม
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(20000000)

# example data
mu = 106  # mean of distribution
sigma = 17  # standard deviation of distribution
x = rng.normal(loc=mu, scale=sigma, size=420)

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Value')
ax.set_ylabel('Probability density')
ax.set_title('Histogram of normal distribution sample: '
             fr'$\mu={mu:.0f}$, $\sigma={sigma:.0f}$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

size = 0.3
vals = np.array([[60., 30.], [36., 41.], [27., 9.]])

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap([1, 2, 5, 6, 9, 10])

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Pie plot')
plt.show()
```
