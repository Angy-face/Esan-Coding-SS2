```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('/content/salecovid.csv')
```

```python
df
```

```python
sns.scatterplot(data = df,x ='Confirmed',y='nettotal')
```

```python
sns.scatterplot(data = df,x ='Month',y='nettotal')
```

```python
features = ['Month','Confirmed']
x = df[features]
y = df['nettotal']
x,y
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x,y)
```

```python
y_pred = linear_reg.predict(x)
y_pred
```

```python
msa = mean_absolute_error(y,y_pred)
msa
```

```python
linear_reg.intercept_
```

```python
plt.scatter(df.Month,y,color='b')
plt.plot(df.Month,y_pred,color='r')
plt.xlabel('Month')
plt.ylabel('nettotal')
plt.title('Linear Regression')
plt.show()
```

```python
print(df.corr(method = 'pearson'))
print(linear_reg.score(x,y))
mean_absolute_error(y,y_pred),mean_squared_error(y,y_pred)
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,shuffle = False)
```

```python
X_train
```

```python
y_train
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)
```

```python
plt.scatter(X_train.Month,y_train,color='b')
plt.plot(X_train.Month,linear_reg.predict(X_train),color='r')
plt.xlabel('Month')
plt.ylabel('nettotal')
plt.title('Linear Regression')
plt.show()
```

```python
y_pred = linear_reg.predict(X_test)
y_pred
```

```python
mse =  mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
mse,mae
```
